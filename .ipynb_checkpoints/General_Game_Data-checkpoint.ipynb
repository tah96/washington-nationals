```python
from bs4 import BeautifulSoup as bs
import requests
import re
from splinter import Browser
import pandas as pd
import itertools
```

```python
#Define Executable Path for ChromeDriver to use for web scraping

executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Define base endpoint URL. Other url's will be built off this endpoint.

baseball_reference_main = 'https://www.baseball-reference.com'
nats_schedule_endpoint = '/teams/WSN/2021-schedule-scores.shtml'
nats_schedule_page = baseball_reference_main + nats_schedule_endpoint
browser.visit(nats_schedule_page)
```

```python
#Use Beautiful Soup HTML.parser to extract relvant details about news articles

schedule_html = browser.html
schedule_soup = bs(schedule_html, 'html.parser')
schedule_soup
```

```python
# Extract list of games

game_listings = schedule_soup.find('table', class_="sortable stats_table now_sortable", id="team_schedule").find('tbody').find_all('tr',class_=lambda x: x != 'thead')
game_listings
```

```python
test = game_listings[57]
test
```

```python
boxscore_braves = test.find_all('td')[1]
boxscore_braves_endpoint = boxscore_braves.a['href']
boxscore_braves_endpoint
```

```python
boxscore_url = baseball_reference_main + boxscore_braves_endpoint
browser.visit(boxscore_url)
box_html = browser.html
box_soup = bs(box_html, 'html.parser')
```

```python
box_soup
```

```python
score_metadata = box_soup.find('div', class_ = 'scorebox_meta').find_all('div')
```

```python
score_metadata
```

```python
game_meta_data = []

for datapoint in score_metadata[:5]:
    meta_contents = datapoint.contents
    if len(meta_contents) > 1:
        key_value_split = meta_contents[1].split(": ")[-1]
        game_meta_data.append(key_value_split)
    else:
        key_value_split = meta_contents[0].split(": ",1)[-1]
        game_meta_data.append(key_value_split)
        
game_meta_data
```

```python
inning_data_headers_raw = box_soup.find('table', class_ = 'linescore nohover stats_table no_freeze').find('thead').find_all('th')
```

```python
inning_data_headers_raw
```

```python
inning_data_headers_text = []

inning_data_headers_text.append('TEAM')
inning_data_headers_text.append('HOME/AWAY')

for header in inning_data_headers_raw:
    header_value = header.text
    if header_value not in ['R','H','E','\xa0']:
        if float(int(header_value)) <= 9:
            inning_data_headers_text.append('INN' + header_value)
        else:
            inning_data_headers_text.append('EXTRAS')
    else:
        inning_data_headers_text.append(header_value)

try:
    while True:
        inning_data_headers_text.remove('\xa0')
except ValueError:
    pass
    
inning_data_headers_text_final = []

for clean_header in inning_data_headers_text:
    if clean_header not in inning_data_headers_text_final:
        inning_data_headers_text_final.append(clean_header)
        
inning_data_headers_text_final
```

```python
inning_by_inning = box_soup.find('table', class_ = 'linescore nohover stats_table no_freeze').find('tbody').find_all('tr')
```

```python
inning_by_inning
```

```python
inning_info_by_team = []

for index, team in enumerate(inning_by_inning):
    data = team.find_all('td')
    inning_stats = []
    for stat in data:
        try:
            inning_data_point = stat.contents[0].text
            inning_stats.append(inning_data_point)
        except AttributeError:
            inning_data_point = stat.contents[0]
            inning_stats.append(inning_data_point)
    if index == 0:
        inning_stats.insert(2,'AWAY')
    else:
        inning_stats.insert(2,'HOME')
    try:
        while True:
            inning_stats.remove('\n')
    except ValueError:
        pass
    
    inning_info_by_team.append(inning_stats)

inning_info_by_team
```

```python
team_test = inning_info_by_team[0]
list_count = len(team_test)
traditional_inning_plus_teams_length = 11
total_dataset_traditional_length = 14
cleaned_list = []
if list_count > total_dataset_traditional_length:
    combined_cleaned_team_lists = []
    for team in inning_info_by_team:
        cleaned_list = []
        innings_to_9 = team[:traditional_inning_plus_teams_length]
        for inning in innings_to_9:
            cleaned_list.append(inning)
        additional_data = team[traditional_inning_plus_teams_length:]
        extra_inning_index_pinpoint = list_count - total_dataset_traditional_length
        extra_inning_data = additional_data[extra_inning_index_pinpoint:]
        extra_runs = additional_data[:extra_inning_index_pinpoint]
        total_scored_in_extras =+ 0
        for inning_runs in extra_runs:
            int_runs = int(inning_runs)
            total_scored_in_extras = total_scored_in_extras + int_runs
        cleaned_list.append(total_scored_in_extras)
        for RHE in extra_inning_data:
            cleaned_list.append(RHE)
        combined_cleaned_team_lists.append(cleaned_list)

combined_cleaned_team_lists
    
        
```

```python
game_dataframe = pd.DataFrame(combined_cleaned_team_lists, columns = inning_data_headers_text_final)
```
