```python
import pandas as pd
```

```python
park_factors_raw = pd.read_csv('park_factors_2020.csv')
```

```python
park_factors_raw
```

```python
team_df = pd.read_csv('team_name_lookup.csv')
team_df
```

```python
partial_team_names = park_factors_raw['Team'].to_list()
team_identification_numbers = []

for team in partial_team_names:
    team_id = team_df.loc[team_df['team_full'].str.contains(team),'team_id'].iloc[0]
    team_identification_numbers.append(team_id)
```

```python
park_factors_raw['Team'] = team_identification_numbers
```

```python
park_factors_raw
```

```python
park_factors_raw.to_csv('park_factors_teamidadj', index = False)
```
