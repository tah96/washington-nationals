```python
import pandas as pd
```

```python
league_pitch_raw = pd.read_csv('LeaguePitchStats.csv')
```

```python
league_pitch_raw
```

```python
team_df = pd.read_csv('team_name_lookup.csv')
team_df
```

```python
team_names = league_pitch_raw['Tm'].to_list()
team_identification_numbers = []

for team in team_names:
    team_id = team_df.loc[team_df['team_full']==team,'team_id'].iloc[0]
    team_identification_numbers.append(team_id)
```

```python
league_pitch_raw['Tm'] = team_identification_numbers
```

```python
league_pitch_raw
```

```python
league_pitch_raw.to_csv('league_pitch_stats_teamidadj', index = False)
```
